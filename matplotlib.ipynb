```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,5,11)
y=x**4
plt.plot(x,y)
plt.title('Hello')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

## Subplots

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(x,y)
```

## Obj Oriented Way

```python
fig=plt.figure(figsize=(12,8), dpi=100)
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis=fig.add_axes([0.3,0.4,.3,.3])
axis.plot(x,y)
```

## Types of Plot

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000), 100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

## Saving plots

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,.8,.8])
ax.plot(x,y)
plt.savefig("basicplot.pdf")
```
